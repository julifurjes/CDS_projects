## Import packages

```python
import spacy
nlp = spacy.load("en_core_web_sm")
import os, glob
import nltk
import numpy as np
import re
import pandas as pd

#pip install -U pip setuptools wheel
#pip install -U spacy
#python -m spacy download en_core_web_sm
```

## Read the files

```python
# assign directory
folder_path = 'in/USEcorpus'

corpus = [""] * 1497
i = 0

for dirpath, dirnames, files in os.walk(folder_path):
    print(f'Found directory: {dirpath}')
    for file_name in files:
        print(file_name)
        with open(os.path.join(dirpath, file_name), "r", encoding="latin-1") as f:
            corpus[i] = f.read()
    if(i < 1497):
        i = i + 1
```

```python
i = 0

for text in corpus:
    corpus[i] = re.sub('<.*?>', '', text)
    corpus[i] = re.sub('[\n\t]', ' ', corpus[i])
    i = i + 1
```

## Creating corpus

```python
# creating empty lists for the tokens
doc = [""] * 1497

# creating a doc object
for i in range(1,1497):
    doc[i] = nlp(corpus[i])
```

## Relative frequency

### Nouns

```python
noun_count = 0
for doc_units in doc:
    for token in doc_units:
        if token.pos_ == "NOUN":
            noun_count += 1

print(noun_count)

relative_freq = round((noun_count/len(doc)) * 10000, 2)
print('Relative frequency of adjectives:', relative_freq)
```

### Verbs

```python
verb_count = 0
for doc_units in doc:
    for token in doc_units:
        if token.pos_ == "VERB":
            verb_count += 1

print(verb_count)

relative_freq = round((verb_count/len(doc)) * 10000, 2)
print('Relative frequency of adjectives:', relative_freq)
```

### Adjectives

```python
adjective_count = 0
for doc_units in doc:
    for token in doc_units:
        if token.pos_ == "ADJ":
            adjective_count += 1

print(adjective_count)

relative_freq = round((adjective_count/len(doc)) * 10000, 2)
print('Relative frequency of adjectives:', relative_freq)
```

### Adverbs

```python
adverb_count = 0
for doc_units in doc:
    for token in doc_units:
        if token.pos_ == "ADV":
            adverb_count += 1

print(adverb_count)

relative_freq = round((adverb_count/len(doc)) * 10000, 2)
print('Relative frequency of adjectives:', relative_freq)
```

## Total number of unique PER, LOC, ORGS

### Creating entities

```python
# create empty list
entities_words = []
entities_labels = []
entities = []

# add each entity to list
for doc_units in doc:
    nlp_doc_units = nlp(doc_units)
    for ent in nlp_doc_units.ents:
        entities_words.append(ent.text)
        entities_labels.append(ent.label_)
        entities.append(ent.text)
        
print(set(entities_words))
print(set(entities_labels))
```

### PER

```python
# creating empty list
per_list = [""]

for doc_units in doc:
    nlp_doc_units = nlp(doc_units)
    for ent in nlp_doc_units.ents:
        if(ent.label_ == 'PERSON'):
            per_list.append(ent.text)

print("Number of unique PERs:", len(set(per_list)))
```

### LOC

```python
# creating empty list
loc_list = [""]

for doc_units in doc:
    nlp_doc_units = nlp(doc_units)
    for ent in nlp_doc_units.ents:
        if(ent.label_ == 'LOC'):
            loc_list.append(ent.text)
    
print("Number of unique LOCs:", len(set(loc_list)))
```

### ORGS

```python
# creating empty list
org_list = [""]

for doc_units in doc:
    nlp_doc_units = nlp(doc_units)
    for ent in nlp_doc_units.ents:
        if(ent.label_ == 'ORG'):
            org_list.append(ent.text)
    
print("Number of unique ORGs:", len(set(org_list)))
```

## Saving the tables

```python
i = 0

noun_count = 0
verb_count = 0
adj_count = 0
adv_count = 0

per_list = [""]
loc_list = [""]
org_list = [""]

for dirpath, dirnames, files in os.walk(folder_path):
    table = pd.DataFrame(columns=['Filename', 'RelFreq NOUN', 'RelFreq VERB', 'RelFreq ADJ', 'RelFreq ADV', 'Unique PER', 'Unique LOC', 'Unique ORG'])
    for file_name in files:
        with open(os.path.join(dirpath, file_name), "r", encoding="latin-1") as f:
            text = f.read()
            text = nlp(text)
            for token in text:
                if token.pos_ == "NOUN":
                    noun_count += 1
                if token.pos_ == "VERB":
                    verb_count += 1
                if token.pos_ == "ADJ":
                    adj_count += 1
                if token.pos_ == "ADV":
                    adv_count += 1
            noun_freq = round((noun_count/len(text)) * 10000, 2)
            verb_freq = round((verb_count/len(text)) * 10000, 2)
            adj_freq = round((adj_count/len(text)) * 10000, 2)
            adv_freq = round((adv_count/len(text)) * 10000, 2)
            for ent in text.ents:
                if(ent.label_ == 'PERSON'):
                    per_list.append(ent.text)
                if(ent.label_ == 'LOC'):
                    loc_list.append(ent.text)
                if(ent.label_ == 'ORG'):
                    org_list.append(ent.text)
            table.loc[i] = [file_name, noun_freq, verb_freq, adj_freq, adv_freq, len(set(per_list)), len(set(loc_list)), len(set(org_list))]
        i = i + 1
        noun_count = 0
        verb_count = 0
        adj_count = 0
        adv_count = 0
        per_list = [""]
        loc_list = [""]
        org_list = [""]
    table.to_csv('out/' + dirpath[-2:] + '.csv', index=False)
    i = 0
```
